```python
# column date
# changing 0 to 1 in the month and day columns
df.loc[(df['imonth'] == 0, 'imonth')] = 1
df.loc[(df['iday'] == 0), 'iday'] = 1
```

```python
# create new column date based on iyear, imonth and iday column values
df.insert(1, 'date', '')
```

```python
from datetime import date
DATE = []
for y,m,d in zip(df.iyear, df.imonth, df.iday):
    DATE.append(date(y,m,d))
    
df['date'] = DATE

df = df.drop(columns = ['iyear','imonth','iday'])
```

```python
df.head()
```

```python
pd.to_datetime(df['date'], format= '%Y-%m-%d')
```

```python
print(df.columns.tolist())
```

```python
df = df.replace({'Unknown':'', -99.0 : '', -9.0 : ''}, regex=True)
```

```python
df.rename(columns={'country_txt':'Country','region_txt':'Region','attacktype1_txt':'Attack_type','target1':'Target',
                   'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                   'weaptype1_txt':'Weapon_type','success':'Success','motive':'Motive'}, inplace=True)
```

```python
new_df = df[['eventid','date','Country','Region','city','latitude','longitude','Attack_type','Killed',
             'Wounded','Target','Summary','Group','Target_type','Weapon_type','Success','Motive']]
```

```python
new_df['casualties'] = new_df['Killed'] + new_df['Wounded']
#make new column
```
